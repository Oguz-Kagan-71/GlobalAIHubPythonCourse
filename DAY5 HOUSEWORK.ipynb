```python
import random
 
# BAZI KISIMLARDA İNTERNETTEN ARŞTIRARAK YARDIM ALDIM.
 
ad = input("İsminiz ? ")

 
print("Hoşgeldin ", ad)
 
kelimeListesi = ['python', 'bilgisayar', 'bilim', 'programlama', 
         'python', 'matematik', 'oyun','pandemi','eğlence','derbeder','hüzün'] 
 
# random kelime seçmeliyiz
kelime = random.choice(kelimeListesi)
 
 
print("harf tahmin et!")
 
tahminler = ''
 
# hak sayısı
turns = 8
 
 
while turns > 0:
     
    # her dönüşte hata 0lanıyor
    hata = 0
     
    
    for char in kelime: 
         
       
        if char in tahminler: 
            print(char)
             
        else: 
            print("_")
             
            # her hatada artırıyoruz
            hata += 1
             
 
    if hata == 0:
        
        print("Kazandın") 
         
        
        print("Kelime: ", kelime) 
        break
     
    
    tahmin = input("tahmin et:")
     
    
    tahminler += tahmin 
     
    
    if tahmin not in kelime:
         
        turns -= 1
         
        print("yanlış")
         
        print("Hakkın", + turns)
         
        if turns == 0:
            print("KAYBETTİN")
```
